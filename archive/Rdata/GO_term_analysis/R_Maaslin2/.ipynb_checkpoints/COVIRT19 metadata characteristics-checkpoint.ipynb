Set the Working Directory

```r
setwd('/home/jovyan/work/Aagaard_Raid3/microbial/GO_term_analysis/')
```

load the necessary libraries

```r
library(tidyverse,quietly = T)
library(dplyr, quietly = T)
library(mosaic, quietly = T)
library(ggpubr, quietly = T)
library(ggsci,quietly = T)
```

Import the metadata

```r
bio_bac_sam<-as.data.frame(read.table("Combined_BALF_GO_Terms_metadata2.txt",header = T, sep = "\t",row.names = 1))
```

Remove all the unkown samples and control negatives

```r
bio_bac_sam<-bio_bac_sam%>%
    filter(case!="Control_Neg")%>%
    filter(case!="Control_Unknown")
```

Do some regex to fix some error in the metadata names

```r
rownames(bio_bac_sam)<-rownames(bio_bac_sam)%>%str_replace_all("NC1_SRR7796663", "NC1.SRR7796663")
bio_bac_sam$accession<-rownames(bio_bac_sam)
```

```r
bio_bac_sam$outcome<-bio_bac_sam$outcome%>%
str_replace_all("recovered", "Recovered")%>%
str_replace_all("deceased","Deceased")%>%
str_replace_all('stabilized',"Stabilized")
```

```r
bio_bac_sam$sex<-bio_bac_sam$sex%>%
str_replace_all("M", "male")%>%
str_replace_all("F", "female")%>%
str_replace_all("na", "")
```

make some tables with the counts and proportions of the datasets

## Cases

```r
case<-tally(case~case,bio_bac_sam,format = "count")
case_perc<-tally(case~case,bio_bac_sam,format = "percent", ,options(digits=3))
cbind(case,case_perc)
```

## Publications

```r
pub<-tally(publication~case,bio_bac_sam,format = "count")
pub_perc<-tally(publication~case,bio_bac_sam,format = "percent", ,options(digits=1))
cbind(pub,pub_perc)
```

## Sex

```r
sex<-tally(sex~case,bio_bac_sam,format = "count")
sex_perc<-tally(sex~case,bio_bac_sam,format = "percent")
sex_perc<-tally(sex~case,bio_bac_sam,format = "percent", ,options(digits=3))
cbind(sex,sex_perc)
```

## Smoking status

```r
smoking_status<-tally(smoking_status~case,bio_bac_sam,format = "count")
smoking_status_perc<-tally(smoking_status~case,bio_bac_sam,format = "percent")
smoking_status_perc<-tally(smoking_status~case,bio_bac_sam,format = "percent", ,options(digits=3))
cbind(smoking_status,smoking_status_perc)
```

## Reads

```r
reads<-tally(reads~case,bio_bac_sam,format = "count")
reads_perc<-tally(reads~case,bio_bac_sam,format = "percent")
reads_perc<-tally(reads~case,bio_bac_sam,format = "percent", ,options(digits=3))
cbind(reads,reads_perc)
```

## Origin

```r
collection_location<-tally(collection_location~case,bio_bac_sam,format = "count")
collection_location_perc<-tally(collection_location~case,bio_bac_sam,format = "percent")
collection_location_perc<-tally(collection_location~case,bio_bac_sam,format = "percent", ,options(digits=3))
cbind(collection_location,collection_location_perc)
```
